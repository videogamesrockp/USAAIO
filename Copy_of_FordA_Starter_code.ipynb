# Part 3.1.1.

Time complexity would be $O(T^2 d)$, as there are $T^2$ dot products to compute with $O(d)$ time complexity to calculate the logits for softmax for, when we do softmax for each row. This is the most computationally intensive step.

# Part 3.1.2.
$$ o_t = \sum_{i=0}^{T-1}\phi(q_t)^⊤ \phi(k_i) v_i$$

The attention output vector $o_t$ is updated with scaled sums weighted based on similarity between the queries and keys, multiplied by the value vector corresponding to the key.

# Part 3.1.3.

$$O = \frac{\phi(Q)\phi(K)^⊤}{∑\phi(Q)\phi(K)^⊤} V $$

The multiplication of $ϕ(Q)$ and $ϕ(K)^⊤$ effectively computes a matrix of how similar each query is to the key. Then this is scaled like softmax. This is then efficiently multiplied by the matrix of value vectors to effectively yield an output matrix where the rows are the output vectors.

# Part 3.1.4.

Assuming $ϕ(Q)$ and $ϕ(K)$ have been computed, the time complexity to find $O$ would be $O(T^2 d)$ as it is simply a matter of multiplying the matrices together to get $O$, which takes T^2 d multiplications.

# Part 3.1.5.

$$\text{exp}\left(\frac{q_i^⊤ k_j}{\sqrt d}\right) = \sum_{r=0}^∞ \frac{(q_i^⊤ k_j/\sqrt d)^r}{r!} = \sum_{r=0}^∞ \frac{(q_i^⊤ k_j)^r}{r!d^{r/2}}$$

This sum would have to require an infinite number of features in the new projections of q and k in order to represent this sum, which cannot be simplified because of the r! in the denominator.

# Part 3.1.6.

$$\exp(\frac{x}{\sqrt d}) ≈ 1 + \frac{x}{\sqrt d} + \frac{x^2}{2d}$$
We want $\phi(x)\phi(x)^⊤$ to equal this approximation, assuming x is $x^\top x$. Therefore, we want

$$\phi(x) = \begin{bmatrix} 1 \\ \frac{x_1}{d^{1/4}} \\ \vdots \\ \frac{x_1^2}{\sqrt {2d}} \\ \vdots \end{bmatrix}$$

# Part 3.2.1.

This relative directional positional encoding is more appropriate than absolute positional encoding when we are processing pairs of tokens and prioritize the relational distance between two tokens in order to determine how strong their relationship is. Using relative directional positional encodings, we can make the queries and keys of closer tokens more similar. This allows attending tokens to prioritize tokens that are closer.

#Part 3.2.2.

j - i can range from -T+1 to T-1. j can be the first and i can be the last, or the other way around. Either way, they are T-1 tokens apart. Every other case, j-1 lies within these extremes.

#Part 3.2.4.

The relative positional encodings can be incorporated by adding them to the keys prior to performing the feature map functions. That way it can be added directly, and since closer query and key vectors prior to the feature map functions should be closer after the feature map function, it will still function as intended and bring them closer in the projected space.
