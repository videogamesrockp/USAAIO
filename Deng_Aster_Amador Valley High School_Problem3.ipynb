# **Problem 3, Part 2**

$\color{green}{\text{### WRITE YOUR SOLUTION HERE ###}}$

1. Since we have 3 rows, we can write A as a decomposition of the form in the problem with only

$$ I = 3 $$

2. For $i ∈ \{0, 1, 2\}$, we set $\mathbf{u}^{(i)}$ and $\mathbf{v}^{(i)}$ as

$i = 0$

$$
\mathbf{u}^{(0)} = \begin{bmatrix}
1 \\
0 \\
0
\end{bmatrix}
$$

$$
\mathbf{v}^{(0)} = \begin{bmatrix}
0 & 0 & 1 & -1
\end{bmatrix}
$$

$i = 1$

$$
\mathbf{u}^{(1)} = \begin{bmatrix}
0 \\
1 \\
0
\end{bmatrix}
$$

$$
\mathbf{v}^{(1)} = \begin{bmatrix}
0 & 0 & 0 & 1
\end{bmatrix}
$$

$i = 2$

$$
\mathbf{u}^{(2)} = \begin{bmatrix}
0 \\
0 \\
1
\end{bmatrix}
$$

$$
\mathbf{v}^{(2)} = \begin{bmatrix}
1 & 2 & 0 & 0
\end{bmatrix}
$$


By setting it like this, the decomposition will really just be the summation of 3 matrices that each contain one non-zero row of the original matrix $\mathbf{A}$.
