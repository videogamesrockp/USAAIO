# **3.1.1**

$$ x_2 = a_2^{1/2} x_{1} + (1-a_2)^{1/2} \epsilon_2 = (a_2)^{1/2}((a_1)^{1/2} x_0 + (1-a_1)^{1/2} \epsilon_1) + (1-a_2)^{1/2} \epsilon_2$$
This is in the form of some $A_2$ multiplied by $x_0$ and the sum of some Gaussian noise components, which sums to also be Gaussian

Now, we find $A_2$,
$$ A_2 = (a_2 a_1)^{1/2}$$

Additionally, $ B_2 = (a_2)^{1/2}(1-a_1)^{1/2} + (1-a_2)^{1/2}$ because we simply sum the two $ \epsilon_1$ and $ \epsilon_2$ that are standard Gaussian distribution.

The variance of the sum of the components of $\epsilon_1$ and $\epsilon_2$ is the sum of their variances, and becomes $2$.

The mean of the Gaussian is still 0.

Therefore, the conditional distribution of $x_2$ given $x_0$ is Gaussian, and has mean of $x_0$ and the variance is $ (a_2) (1 - a_1) + (1 - a_2) $

# **3.1.2**

$$ x_t = a_t^{1/2} x_{t-1} + (1-a_t)^{1/2} \epsilon_t $$
If $x_{t-1}$ can be written in the form $A_{t-1} x_0 + B_{t-1} ̅{\epsilon_{t-1}} $, then we have

$$ x_t = a_t^{1/2} (A_{t-1} x_0 + B_{t-1} ̅{\epsilon_{t-1}}) + (1-a_t)^{1/2} \epsilon_t$$

The sum of $a_t^{1/2} B_{t-1} ̅{\epsilon_{t-1}} $ and $(1-a_t)^{1/2} \epsilon_t$ would be a Gaussian distribution because both are Gaussian distribution. Additionally, since it is known that $x_1$ and $x_2$ satisfy the prerequisite for this rule, we therefore have proof that all $x_t$ can be written in this form by induction.

$$ A_t = \Pi_{i=1}^t a_i^{1/2} = ̅{a_t}^{1/2}$$

as it is equal to the previous $A_{t-1}$ multiplied by $a_t$, which itself is a summation of all previous {a_t}. Thus this is derived similarly through the recursive rule.

$$ B_t = a_t^{1/2} B_{t-1} + (1-a_t)^{1/2} = a_t^{1/2}(a_{t-1}^{1/2} B_{t-2} + (1-a_{t-1})^{1/2}) + (1-a_t)^{1/2} = a_t^{1/2}(a_{t-1}^{1/2}(a_{t-2}^{1/2} + (1-a_{t-2}^{1/2} + (1-a_{t-1})^{1/2}) + (1-a_t)^{1/2} $$
which goes on infinitely down to the base $B_2$

The conditional distribution of x_t given x_0 is Gaussian because it is a sum of a Gaussian term and a constant. The mean would be $ ̅{a_t}^{1/2} x_0$, and the variance would be $t$ as it is the sum of t Gaussian distributions.
