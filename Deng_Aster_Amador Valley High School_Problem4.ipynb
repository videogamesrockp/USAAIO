Using the quotient rule, the derivative becomes:

$$
\frac{d \tanh x}{dx} = \frac{\frac{d (e^x - e^{-x})}{dx} × (e^x + e^{-x}) - \frac{d (e^x + e^{-x})}{dx} × (e^x - e^{-x})} {(e^x + e^{-x})^2}
$$

Solving the derivatives yields

$$
\frac{d \tanh x}{dx} = \frac{(e^x+e^{-x})^2 - (e^x-e^{-x})^2}{(e^x + e^{-x})^2}
$$

Expanding the numerator and canceling out the terms gives

$$
\frac{d \tanh x}{dx} = \frac{4}{(e^x+e^{-x})} = (\frac{2}{(e^x+e^{-x})})^2
$$

So the derivative is

$$
\frac{d \tanh x}{dx} = \mathrm{sech}^2(x)
$$
